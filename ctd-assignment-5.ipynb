## Task 1: Data Selection

```python
# Part 1
data1 = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 30],
    'Salary': [50000, 60000, 70000, 80000, 55000]
}

data2 = {
    'Name': ['Frank', 'Grace', 'Helen', 'Ian', 'Jack'],
    'Age': [28, 33, 35, 29, 40],
    'Salary': [52000, 58000, 72000, 61000, 85000]
}

data3 = {
    'Name': ['Frank', 'Helen', 'Ian', 'Hima', 'Chaka'],
    'Age': [17, 93, 12, 57, 106],
    'Favorite Color': ['blue', 'pink', 'burgundy', 'red', 'turquoise']
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)

display(df1)
display(df2)
display(df3)

# Part 2
print(df1["Name"])
print(df1[["Name", "Salary"]])
print(df1.iloc[0:3])
```

## Task 2: Data Aggregation

```python
df1_by_age = df1.groupby("Age").agg(mean = ("Salary", "mean"), salary_sum = ("Salary", "sum"), count = ("Salary", "count"))
display(df1_by_age)
```

## Task 3: Merging and Joining DataFrames

```python
# Part 1
warnings.filterwarnings('ignore')
df_1_3_merged = pd.merge(df1, df3, on = "Name", how = "outer", suffixes=['_left','_right'])
display(df_1_3_merged)

df_1_3_merged = df_1_3_merged.fillna({"Salary" : 15000, "Favorite Color" : "yellow"})
display(df_1_3_merged)

df_1_3_merged['Age'] = np.where(df_1_3_merged['Age_left'].notna(),df_1_3_merged['Age_left'], df_1_3_merged['Age_right'])
display(df_1_3_merged)

df_1_3_merged.drop(labels = ["Age_left", "Age_right"], axis = 1, inplace = True)
display(df_1_3_merged)

# Part 2
df1_b = df1.copy().set_index("Name")
df3_b = df3.copy().set_index("Name")
joined_df1_df3 = df1_b.join(df3_b, on = "Name", how = "outer", lsuffix='_left', rsuffix='_right')
display(joined_df1_df3)
```

## Task 9: Data Wrangling a Kaggle Dataset

```python
football_results = pd.read_csv("/kaggle/input/international-football-results-from-1872-to-2017/results.csv")
print(football_results.head())

results_1 = football_results[["home_team", "away_team", "home_score", "away_score", "date"]]
print(results_1.head())

results_2 = results_1.rename(columns = {"home_team" : "team", "away_team" : "opponent", "home_score" : "points_for", "away_score" : "points_against"})
print(results_2.head())

results_3 = results_1.rename(columns = {"away_team" : "team", "home_team" : "opponent", "away_score" : "points_for", "home_score" : "points_against"})
# print(results_3.head())
print(results_3.head())

football_results = pd.concat([results_2, results_3], ignore_index = True)
print(football_results.head())

points_against = football_results.groupby("team")["points_against"].mean()

sorted_points_against = points_against.sort_values(ascending = False)
print(sorted_points_against.head(10))
```
