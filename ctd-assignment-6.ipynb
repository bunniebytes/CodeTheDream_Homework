## Task 1: Practice Pivot Tables

```python
warnings.filterwarnings('ignore')
ecommerce = pd.read_csv("/kaggle/input/ecommerce-consumer-behavior-analysis-data/Ecommerce_Consumer_Behavior_Analysis_Data.csv")
display(ecommerce.head())

ecommerce["Purchase_Amount"] = pd.to_numeric(ecommerce["Purchase_Amount"].str.replace("$", "", regex = False), errors = "coerce")

buying_patterns = pd.pivot_table(ecommerce, index = "Purchase_Category", columns = ["Gender", "Income_Level"], values = "Purchase_Amount", aggfunc = "sum")
display(buying_patterns)

demographics = pd.pivot_table(ecommerce, index = ["Income_Level", "Education_Level"], columns = "Marital_Status", values = "Customer_ID", aggfunc = "count")
display(demographics)
```

## Task 2: Practice apply()

```python
jobs = pd.read_csv("/kaggle/input/ai-powered-job-recommendations/job_recommendation_dataset.csv")
display(jobs.head(10))

def is_entry_level(row):
    skills = row["Required Skills"].replace(" ", "").split(",")
    if row["Experience Level"] == "Entry Level" and row["Salary"] >= 70000 and "SQL" in skills and "Python" in skills:
        return "Yes"
    return "No"

jobs["Check These Out"] = jobs.apply(is_entry_level, axis = 1)

my_jobs = jobs[jobs["Check These Out"] == "Yes"]
display(my_jobs.head(10))
```

## Task 3: Handling Missing Data

```python
df = pd.read_csv("/kaggle/input/code-the-dream-lesson-5/fake_data.csv")
display(df)

df1 = df.dropna()
display(df.info())
display(df1.info())

values = {"Name" : "Unknown", "Age" : df["Age"].mean(), "Salary" : df["Salary"].median(), "Join Date" : "2020-01-01"}
df.fillna(value = values, inplace = True)
display(df)

df2 = df.dropna(ignore_index = True).astype({"Age" : int})
display(df2)
```

## Task 4: Data Transformation

```python
df3 = pd.read_csv("/kaggle/input/code-the-dream-eclipses/eclipses.csv", sep = "|")
display(df3.info())
display(df3.head())

df3["Date"] = pd.to_datetime(df3["Date"], errors= "coerce")
display(df3.head(20))
```

## Task 5: Removing Duplicates

```python
display(df3.info())
duplicate_series = df3.duplicated()
print(duplicate_series[duplicate_series == True].head(10))
print(duplicate_series.value_counts())
df3.drop_duplicates(inplace = True)
display(df3.info())
```

## Task 6: Handling Outliers

```python
outliers = (df2["Age"] > 100) | (df2["Age"] < 0)
median_age = int(df2[(df2["Age"] < 100) & (df2["Age"] > 0)]["Age"].median())

df2.loc[outliers, "Age"] = median_age
display(df2)
```

## Task 7: Standardizing Data

```python
df2["Name"] = df2["Name"].str.lower().str.strip()
display(df2)

replace_map = {"NYC" : "New York City", "LA" : "Los Angeles"}
df2["City"] = df2["City"].replace(replace_map)
display(df2)
```

## Task 8: Validating Data Ranges

```python
invalid_age = (df2["Age"] < 18) | (df2["Age"] > 65)
nan = np.nan
df2.loc[invalid_age, "Age"] = nan
display(df2)

df2["Age"] = df2["Age"].fillna(median_age)
display(df2)
```
